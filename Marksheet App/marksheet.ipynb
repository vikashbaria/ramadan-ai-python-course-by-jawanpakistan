```python
print("===== Python Marksheet =====\n")

english = int(input("Enter Your English Marks: "))
urdu = int(input("Enter Your Urdu Marks: "))
math = int(input("Enter Your Math Marks: "))
pakstudies = int(input("Enter Your Pakistan Studies Marks: "))
sindhi = int(input("Enter Your Sindhi Marks: "))

subjects = {
    "English": english,
    "Urdu": urdu,
    "Math": math,
    "Pakistan Studies": pakstudies,
    "Sindhi": sindhi
}

total_marks = english + urdu + math + pakstudies + sindhi
max_total = 500  # assuming each subject 100 marks
percentage = total_marks / max_total * 100

if percentage <= 45:
    grade = "F"
elif percentage <= 50:
    grade = "E"
elif percentage <= 60:
    grade = "D"
elif percentage <= 65:
    grade = "C"
elif percentage <= 75:
    grade = "B"
elif percentage <= 85:
    grade = "A"
elif percentage <= 94:
    grade = "A+"
else:
    grade = "Outstanding"

fail_subjects = [sub for sub, mark in subjects.items() if mark < 33]

print("\n===== MARKSHEET =====")
for sub, mark in subjects.items():
    print(f"{sub:<20}: {mark} / 100")

print(f"\nTotal Marks       : {total_marks} / {max_total}")
print(f"Percentage        : {percentage:.2f}%")
print(f"Grade             : {grade}")

if fail_subjects:
    print("Failed Subjects   :", ", ".join(fail_subjects))
else:
    print("Status            : Passed in all subjects")
```
