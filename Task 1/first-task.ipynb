```python
print("Marksheet in python\n")
english = int(input ("Enter Your English Marks"))
urdu = int(input ("Enter Your Urdu Marks"))
math = int(input ("Enter Your Math Marks"))
pakstudies = int(input ("Enter Your Pakistan Studies Marks"))
sindhi = int(input ("Enter Your Sindhi Marks"))

totalmarks = english+urdu+math+pakstudies+sindhi
print("\nTotal Marks of 5 subjects will be : ", totalmarks)
percentage = int(totalmarks/500*100)

if percentage <= 45:
    print("Grade F")
elif percentage <= 50:
    print("Grade E")
elif percentage <= 60:
    print("Grade D")
elif percentage <= 65:
    print("Grade C")
elif percentage <= 75:
    print("Grade B")
elif percentage <= 85:
    print("Grade A")
elif percentage <= 94:
    print("Grade A+")
else:
    print("Better Luck for Next Time")
```
