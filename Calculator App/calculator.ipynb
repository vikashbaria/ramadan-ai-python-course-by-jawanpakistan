```python
print("Biggener Level Calculator\n")
num1 =int(input("Enter Value one "))
num2 =int(input("Enter Value two "))
sign = input("What task you want to perform +, -, x, %: ")
if sign == "+":
    print(num1+num2)
elif sign == "-":
    print(num1-num2)
elif sign == "*":
    print(num1*num2)
elif sign == "%":
    print(num1%num2)
else:
    print("Please add correct sign we only allow +-x%")
```
